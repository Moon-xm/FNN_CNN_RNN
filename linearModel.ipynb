### 课堂示例

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

def forward(x):
    return w*x

def loss(x, y):
    y_pred = forward(x)
    return (y - y_pred)*(y - y_pred)

w_ls = []
mse_ls = []
for w in np.arange(0.0, 4.1, 0.1):
#     print('w=',w)
    l_sum = 0
    for x_val,y_val in zip(x_data, y_data):
        y_pred_val = forward(x_val)
        loss_val = loss(x_val, y_val)
        l_sum += loss_val
#         print('\t', x_val, y_val, y_pred_val, loss_val)
#     print('MSE=',l_sum/3)
    w_ls.append(w)
    mse_ls.append(l_sum/3)

plt.plot(w_ls, mse_ls)
plt.xlabel('w')
plt.ylabel('MSE')
plt.show()
```

### 课堂作业

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x_data = np.array([1.0, 2.0, 3.0])
y_data = np.array([2.0, 4.0, 6.0])
w = np.arange(0.0, 4.1, 0.1)  # 41
b = np.arange(-1.0, 1.1, 0.1)  # 21
W,B = np.meshgrid(w, b)  # 21*41 21*41

def mse(w, b):
    y_pred = w*x_data + b
    mse_val = np.square(y_pred - y_data).sum()/ len(y_data)
    return mse_val

mse_ls = []

for i in range(len(b)):
    mse_ls_temp = []
    for j in range(len(w)):
        mse_ls_temp.append(mse(W[i][j], B[i][j]))
    mse_ls.append(mse_ls_temp)

Z = np.array(mse_ls)

fig = plt.figure()
ax = Axes3D(fig)
ax.set_xlabel('w')
ax.set_ylabel('b')
ax.set_zlabel('MSE')
ax.plot_surface(W, B, Z, cmap='jet')
plt.show()
# y_pred = W*x_data + B
# y_pred
```

### 3D图形的库函数使用

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
```

```python
# 将0-15分为1000份
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)

# 创建3D画布
fig = plt.figure()
ax = Axes3D(fig)

# 绘制图形
ax.plot(xline,yline,zline)  # 3D曲线图
plt.show()
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)  # 生成0-1之间的随机数
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x, y, z, color ='r')  # 3D散点图 color='r'设置散点颜色为红色
plt.show()
```
